```python
# deleting irrelevant columns
new_df=sales_df.drop(columns=['Invoice ID','cogs','gross margin percentage','gross income'], axis=1)
```

```python
new_df
```

```python
# checking for duplicates
new_df.duplicated().sum()
```

```python
new_df
```

```python
# creating new column
new_df['Month']=new_df['Date'].str[0]
```

```python
new_df
```

```python
# finding the total number of sales per gender
sales_per_gender=new_df.groupby('Gender', as_index=False).sum(numeric_only=True)
sales_per_gender
```

```python
plt.figure(figsize=(5,3))
plt.pie( sales_per_gender.Total, labels=sales_per_gender.Gender, autopct='%1.0f%%')
plt.title('Sales Per Gender')
plt.show()
```

#### Sales generated from female customers is slightly higher than that generated from male customers

```python
# finding sales per city
sales_per_city=new_df.groupby('City', as_index=False).sum(numeric_only=True)
sales_per_city=sales_per_city.sort_values('Total')
sales_per_city
```

```python
plt.figure(figsize=(5,3))
plt.bar(sales_per_city['City'], sales_per_city['Total'], width=0.6)
plt.title('Sales Per City')
plt.xlabel('Cities')
plt.ylabel('Total Sales in $')
plt.show()
```

#### Naypyitaw records the highest number of sales followed by Yangon and then Mandalay, however sales recorded from Yangon are slightly higher than Mandalay

```python
# finding the most ordered productline
preferred_productline=new_df.groupby('Product line', as_index=False).sum(numeric_only=True)
```

```python
preferred_productline=preferred_productline.sort_values('Quantity')
```

```python
plt.figure(figsize=(5,3))
plt.bar(preferred_productline['Product line'], preferred_productline['Quantity'], width=0.6, color='orange')
plt.title('Quantity Ordered Per Product line')
plt.xlabel('Product line')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation='vertical')
plt.show()
```

#### Electronic accessories are the most ordered productline with Health and beauty products being the least ordered product line

```python
# finding total sales per product line
TotalRevenue=preferred_productline.sort_values('Total')
plt.figure(figsize=(5,3))
plt.title('Total Sales Per Product line')
plt.bar(TotalRevenue['Product line'], TotalRevenue['Total'], width=0.6)
plt.xticks(rotation='vertical')
plt.show()
```

```python
tax_mean=new_df.groupby('Product line', as_index=False).mean(numeric_only=True)
```

```python
tax_mean=tax_mean.sort_values('Unit price')
```

```python
plt.figure(figsize=(5,3))
plt.barh(tax_mean['Product line'], tax_mean['Unit price'], color='orange')
plt.show() 
```

#### Food and beverages recorded the highest sales although being the second most ordered productline. Health and beauty products recorded the lowest sales which is probably due to the fact the it is the least preferred product line

```python
# finding the sales per each branch
sales_per_branch=new_df.groupby('Branch', as_index=False).sum(numeric_only=True).sort_values('Total')
```

```python
sales_per_branch
```

```python
plt.figure(figsize=(5,3))
plt.barh(sales_per_branch.Branch, sales_per_branch.Total)
plt.title('Total Sales Per Branch')
plt.xlabel('Total Sales in $')
plt.ylabel('Branch')
plt.show()
```

#### Branch C generated the highest sales with Branch B recording the least

```python
# finding the most preferred payment method
mostPreferredPaymentMode=new_df.groupby('Payment')['Payment'].count().plot(kind='barh', figsize=(5,3), color='orange')
```

#### From the bar chart, the most preferred mode of payment is via Ewallet and Cash

```python
customerType_gender=new_df.groupby(['Customer type'], as_index=False).sum(numeric_only=True)
```

```python
customerType_gender
```

```python
plt.figure(figsize=(5,3))
plt.pie(customerType_gender['Total'], labels=customerType_gender['Customer type'], autopct='%1.0f%%')
plt.show()
```

#### Totale sales generated from normal customers are slightly lower than sales generated from members

```python
# finding the average rating per product line
avg_rating=new_df.groupby('Product line', as_index=False).mean(numeric_only=True).round(2).sort_values('Rating', ascending=False)
```

```python
avg_rating
```

```python
plt.figure(figsize=(5,3))
plt.bar(avg_rating['Product line'], avg_rating['Rating'])
plt.title('Rating Per Product line')
plt.xlabel('Product line')
plt.ylabel('Rating')
plt.xticks(rotation='vertical')
plt.show()
```
